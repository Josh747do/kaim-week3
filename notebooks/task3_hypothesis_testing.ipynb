```python
from src.hypothesis_tests import (
    load_data,
    create_bmi_category,
    t_test_groups,
    anova_test,
    chi_square_test,
    interpret_p_value
)

df = load_data("../data/insurance.csv")
df.head()
```

```python
# Hypothesis 1: Smoker vs Non-Smoker

smokers = df[df["smoker"] == "yes"]["charges"]
nonsmokers = df[df["smoker"] == "no"]["charges"]

t_stat, p_val = t_test_groups(smokers, nonsmokers)

print("T-test: Smoker vs Non-Smoker")
print("t-statistic:", t_stat)
print("p-value:", p_val)
print(interpret_p_value(p_val))
```

```python
# Hypothesis 2: Male vs Female

male = df[df["sex"] == "male"]["charges"]
female = df[df["sex"] == "female"]["charges"]

t_stat, p_val = t_test_groups(male, female)

print("T-test: Male vs Female")
print("t-statistic:", t_stat)
print("p-value:", p_val)
print(interpret_p_value(p_val))
```

```python
# Hypothesis 3: Region effect on charges (ANOVA)

f_stat, p_val = anova_test(df, "region")

print("ANOVA: Charges across Regions")
print("F-statistic:", f_stat)
print("p-value:", p_val)
print(interpret_p_value(p_val))
```

```python
df = create_bmi_category(df)
df.head()
```

```python
# Hypothesis 4: BMI Group (High BMI vs Normal BMI)

high_bmi = df[df["bmi_group"] == "high_bmi"]["charges"]
normal_bmi = df[df["bmi_group"] == "normal_bmi"]["charges"]

t_stat, p_val = t_test_groups(high_bmi, normal_bmi)

print("T-test: High BMI vs Normal BMI")
print("t-statistic:", t_stat)
print("p-value:", p_val)
print(interpret_p_value(p_val))
```
